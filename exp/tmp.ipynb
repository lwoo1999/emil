```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
angle = np.random.randn(10000) * 0.1 + 0.5
```

```python
plt.hist(angle)
```

```python
n = np.random.rand(10000)
```

```python
plt.hist(n)
```

```python
sb.histplot(angle)
sb.histplot(angle[n < angle], color="orange")
```

```python
angle_ = angle[n < angle]
```

```python
len(angle_)
```

```python
a = None
b = None

while True:
    a = np.random.randn(4933)
    b = np.random.randn(4933)
    corr = np.corrcoef(a, b)[0, 1]
    if np.abs(corr) < 1e-4:
        break
```

```python
np.corrcoef(a, b)
```

```python
sb.histplot(np.log10(angle_))
```

```python
np.corrcoef(0.16*a + np.log10(angle_), 0.16*b - np.log10(angle_))
```

```python
x = np.linspace(0, 0.3)
y = np.vectorize(lambda x: np.corrcoef(x*a + np.log10(angle_), x*b - np.log10(angle_))[0,1])(x)
plt.plot(x, y)
```
